# Clean routes.sql

In some routes, there's more than one plane inserted, so we're separating per line here.

```python
import pandas as pd
headers = ['airline', 'airline_id', 'source_airport', 'source_airport_id', 'destination_airport', 'destination_airport_id', 'codeshare', 'stops', 'equipment']
routes = pd.read_csv("routes.csv", names=headers)
routes.head()
```

```python
new_lines = pd.DataFrame(columns=headers)
new_lines
```

```python
for i in range(1):
    print(routes.iloc[i])
    if " " in str(routes.iloc[i]['equipment']):
        planes = str(routes.iloc[i]['equipment']).split(' ')
        new_line = routes.iloc[i].copy()
        new_line['equipment'] = planes[0]
        new_lines = new_lines.append(new_line, ignore_index=True)

        for j in range(1, len(planes)):
            new_line = routes.iloc[i].copy()
            new_line['equipment'] = planes[j]
            new_lines = new_lines.append(new_line, ignore_index=True)
    else:
        new_lines = new_lines.append(routes.iloc[i].copy(), ignore_index=True)
        
new_lines
```

```python
new_lines
```

```python
new_lines.to_csv("routes-treated.csv", header=False, index=False)
```
